```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
import re
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,roc_curve,classification_report
nltk.download('stopwords')
from pylab import rcParams
nltk.download('wordnet')
nltk.download('punkt')
import string as st
```

```python
data = pd.read_csv('/content/train - train.csv')
df_test = pd.read_csv('/content/test - test.csv')
```

```python
def preprocess(data):
  def lemmatize(text):
    word_net = WordNetLemmatizer()
    return [word_net.lemmatize(word) for word in text]
  def remove_stopwords(text):
    return [word for word in text if word not in nltk.corpus.stopwords.words('english')]
  def remove_small_words(text):
    return [x for x in text if len(x) > 3 ]
  def tokenize(text):
    text = re.split('\s+' ,text)
    return [x.lower() for x in text]
  def remove_punct(text):
    return ("".join([ch for ch in text if ch not in st.punctuation]))
  data['text'] = data['text'].apply(lambda x: remove_punct(x))
  data['text'] = data['text'].apply(lambda x : tokenize(x))
  data['text'] = data['text'].apply(lambda x : remove_small_words(x))
  data['text'] = data['text'].apply(lambda x : lemmatize(x))
  return data
  
```
